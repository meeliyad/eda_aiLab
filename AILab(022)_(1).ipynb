```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load dataset
def load_dataset(file_path):
    return pd.read_csv(file_path)

df = load_dataset("weather_data.txt")
df.head(10)
```

```python
# Categorical data
# countplot
sns.countplot(data=df, x='weather')
df['date'].value_counts()
```

```python
 # pie chart

df['weather'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
# Numberical Value

# Histogram treat as categorical data distribution of data

plt.hist(df['wind'])
```

```python
# Distplot
# probability density function --> The line is kde
# skweness
sns.distplot(df['wind'])
```

```python
# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier

sns.boxplot(df['wind'])
```

```python
df.describe()
```

```python
# Bar plot (Numerical - Categorical)
  # average wind
sns.barplot(x='weather',y='wind',hue='weather',data=df)
```

```python
# Displot

sns.distplot(df[df['wind']==0]['temp_max'],hist=False)
sns.distplot(df[df['wind']==1]['temp_max'],hist=False)
```

```python
sns.lineplot(x='weather',y='date',data=df.head(10))
```

```python
# HeatMap (Categorical - Categorical)
sns.heatmap(pd.crosstab(df['weather'],df['wind']))
```

```python
# HeatMap (Categorical - Categorical)
sns.heatmap(pd.crosstab(df['wind'],df['weather']))
```

```python
sns.lineplot(x='date',y='temp_max',data=df.head(5))
```

```python
sns.boxplot(x='weather',y='wind',hue='temp_max',data=df.head(30))
```

```python
# Scatter plot(Numerical - Numerical) for the relation between two column
sns.scatterplot(x='temp_max', y='date',hue='weather' ,data=df.head(25))
plt.show()
```
